```python
import pandas as pd
import numpy as np
```

# Compare CPI Score and Happiness Score

### There are only overlapping data for 2015 and 2016

```python
df0 = pd.read_csv('history.csv',engine='python')
df0.head()
```

```python
df1 = pd.read_csv('2015.csv',engine='python')
df1.head()
```

```python
df2 = pd.read_csv('2016.csv',engine='python')
df2.head()
```

## Find the correlation between Happiness Score in 2015 and CPI 2015 Score

```python
df3=df0[['Country','CPI 2015 Score']]
```

```python
df4=df1[['Country', 'Happiness Score']]
```

```python
df5= pd.merge(left = df4, right = df3,how= 'inner',left_on='Country', right_on='Country')
df5.head()
```

```python
#The correlation between Happiness Score in 2015 and CPI 2015 Score: 0.67
corr1 = df5.corr(method = 'pearson')
print(corr1)
```

## Find the correlation between Happiness Score in 2016 and CPI 2016 Score

```python
df6=df0[['Country','CPI 2016 Score']]
```

```python
df7=df2[['Country', 'Happiness Score']]
```

```python
df8= pd.merge(left = df7, right = df6,how= 'inner',left_on='Country', right_on='Country')
df8.head()
```

```python
#The correlation between Happiness Score in 2016 and CPI 2016 Score: 0.68
corr2 = df8.corr(method = 'pearson')
print(corr2)
```
