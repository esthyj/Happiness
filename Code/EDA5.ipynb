```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Find the correlation betwen Happiness score and homicide rate, suicide rate, GDP by each year

```python
df0 = pd.read_csv('suicide.csv',engine='python')
df0.head()
```

```python
df1=df0[['country','year','Intentional homicides ','Suicide mortality rate']]
df1.head()
```

# 2015

```python
df2=df1[df1['year']==2015]
df2.dropna()
```

```python
df3 = pd.read_csv('2015.csv',engine='python')
df3.head()
```

```python
df4=df3[['Country','Happiness Score']]
```

```python
df5= pd.merge(left = df4, right = df2,how= 'inner',left_on='Country', right_on='country')
df5.drop(['year'],axis=1, inplace=True)
df5.head()
```

```python
sns.set(style="darkgrid")
corr = df5.corr()
f, ax = plt.subplots(figsize=(4, 4))
X = sns.heatmap(corr,  cmap='Blues', vmax=1, vmin=-1, center=0,annot=True, fmt='.2f',
                square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot_kws={"size": 15})
```
